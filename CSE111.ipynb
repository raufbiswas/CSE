```
Function Call       Sample Output
(â€˜madamâ€™)           Palindrome
(â€˜helloâ€™)           Not a palindrome
(â€˜nurses runâ€™)      Palindrome
```

`````python
def palin(st):
  s = ''
  for i in st:
    if i != ' ':
      s+=i
  p = s[::-1]
  if p == s:
    print('Palindrome')
  else:
    print('Not a palindrome')
palin('madam')
`````

`````python
#task10
class Student:
    def __init__(self,name,ID):
        self.name = name
        self.ID = ID
    def Details(self):
        return "Name: "+self.name+"\n"+"ID: "+self.ID+"\n"
class CSEStudent(Student):
    def __init__(self,name,ID,semester):
        super().__init__(name,ID)
        self.semester = semester
        self.coursegrade = {}
    def Details(self):
        return f"{super().Details()}Current semester: {self.semester}"
    def addCourseWithMarks(self,*coursewithmarks):
        self.number_of_course = int(len(coursewithmarks)/2)
        for i in range(0,len(coursewithmarks),2):
            if coursewithmarks[i+1] >= 85:
                self.coursegrade[coursewithmarks[i]] = 4.0
            elif 80<=coursewithmarks[i+1]<=84:
                self.coursegrade[coursewithmarks[i]] = 3.3
            elif 70<=coursewithmarks[i+1]<=79:
                self.coursegrade[coursewithmarks[i]] = 3.0
            elif 65<=coursewithmarks[i+1]<=69:
                self.coursegrade[coursewithmarks[i]] = 2.3
            elif 57<=coursewithmarks[i+1]<=64:
                self.coursegrade[coursewithmarks[i]] = 2.0
            elif 55<=coursewithmarks[i+1]<=56:
                self.coursegrade[coursewithmarks[i]] = 1.3
            elif 50<=coursewithmarks[i+1]<=54:
                self.coursegrade[coursewithmarks[i]] = 1.0
            else:
                self.coursegrade[coursewithmarks[i]] = 0.0
    def showGPA(self):
        print(f"{self.name} has taken {self.number_of_course} courses.")
        cgpa = 0
        for key,value in self.coursegrade.items():
            print(f"{key}: {value}")
            cgpa += value
        self.gpa = round(cgpa/self.number_of_course,2)
        print(f"GPA of {self.name} is: {self.gpa}")
Bob = CSEStudent("Bob","20301018","Fall 2020")
Carol = CSEStudent("Carol","16301814","Fall 2020")
Anny = CSEStudent("Anny","18201234","Fall 2020")
print("#########################")
print(Bob.Details())
print("#########################")
print(Carol.Details())
print("#########################")
print(Anny.Details())
print("#########################")
Bob.addCourseWithMarks("CSE111",83.5,"CSE230",73.0,"CSE260",92.5)
Carol.addCourseWithMarks("CSE470",62.5,"CSE422",69.0,"CSE460",76.5,"CSE461",87.0)
Anny.addCourseWithMarks("CSE340",45.5,"CSE321",95.0,"CSE370",91.0)
print("----------------------------")
Bob.showGPA()
print("----------------------------")
Carol.showGPA()
print("----------------------------")
Anny.showGPA()
`````
